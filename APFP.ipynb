```python
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori 
```

```python
store_data = pd.read_csv('C:\\Users\\Vidyajay\\Desktop\\NEURAL NETWORK\\ak.csv',header=None)  
```

```python
store_data.shape
```

```python
store_data.head()
```

```python
records = []  
for i in range(0,25 ):  
    records.append([str(store_data.values[i,j]) for j in range(0, 5)])
```

```python
records
```

```python
association_rules = apriori(records, min_support=0.025, min_confidence=0.5, min_lift=4, min_length=4
                           )  
association_results = list(association_rules) 
```

```python
print(len(association_results)) 
```

```python
association_results
```

```python
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
